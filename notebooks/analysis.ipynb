```python
import pandas as pd
#=========================
# ANALYSIS SECTION
#=========================
df = pd.read_csv("data/clean_sales_data.csv")

#Total sales
total_sales = df['sales'].sum()
print("Total Sales:", total_sales)

#Sales by years
sales_by_year = df.groupby('year')['sales'].sum()
print("Sales by year:", sales_by_year)

#Sales by regions
sales_by_region = df.groupby('region')['sales'].sum()
print("Sales by region:", sales_by_region)

#Sales by Category
sales_by_category = df.groupby('category')['sales'].sum()
print("Sales by category:", sales_by_category)

#Sales by month
sales_by_month =df.groupby('month')['sales'].sum()
print("Sales by Month:", sales_by_month)

#Top 10 Products
top_products = df.groupby('product_name')['sales'].sum().sort_values(ascending=False).head(10)
print("Top products:", top_products)

#Average Shipping days
avg_shipping = df['shipping_days'].mean()
print("Average shipping:", avg_shipping)
```
