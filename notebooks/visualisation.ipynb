```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn style
sns.set(style="whitegrid")

# Create folder to save images
if not os.path.exists("images"):
    os.makedirs("images")

#Load clean dataset
df = pd.read_csv("data/clean_sales_data.csv")

#=========================
# 1 Sales by Years (Bar)
#=========================
sales_by_year =df.groupby('year')['sales'].sum()

#Plot
plt.figure(figsize=(8,5))
ax = sns.barplot(x=sales_by_year.index, y=sales_by_year.values, color="blue")
plt.title("Sales by Year", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Sales ()", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Save the image
plt.savefig("images/sales_by_year_bar.png", dpi=300)
plt.show()

#=========================
# 2 Sales by Years (Line)
#=========================
plt.figure(figsize=(8,5))
ax = sns.lineplot(x=sales_by_year.index, y=sales_by_year.values, marker='o', color='green')
plt.title("Sales Trend by Year", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Sales ()", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Save the image
plt.savefig("images/sales_by_year_line.png", dpi=300)
plt.show()

#=========================
# 3 Sales by Regions
#=========================
sales_by_region =df.groupby('region')['sales'].sum()

plt.figure(figsize=(8,5))
ax = sns.barplot(x=sales_by_region.index, y=sales_by_region.values, color="green")
plt.title("Sales by Region", fontsize=14)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Total Sales ()", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

#Save the image
plt.savefig("images/sales_by_region.png", dpi=300)
plt.show()

#=========================
# 4 Sales by Category
#=========================
sales_by_category =df.groupby('category')['sales'].sum()

plt.figure(figsize=(8,5))
ax = sns.barplot(x=sales_by_category.index, y=sales_by_category.values, color="orange")
plt.title("Sales by Category", fontsize=14)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Total Sales ()", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

#Save the image
plt.savefig("images/sales_by_category.png", dpi=300)
plt.show()

#=========================
# 5 Sales by Month
#=========================
sales_by_month =df.groupby('month')['sales'].sum()

plt.figure(figsize=(8,5))
ax = sns.lineplot(x=sales_by_month.index, y=sales_by_month.values, marker='o', color='purple')
plt.title("Monthly Sales Trend", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales ()", fontsize=12)
plt.xticks(rotation=0)  # Months usually look better horizontal
plt.tight_layout()

#Save the image
plt.savefig("images/sales_by_month.png", dpi=300)
plt.show()
```
